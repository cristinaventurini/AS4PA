## Exercise 1 - vectors and data frames

The table linked gives the volume, area, length and maximum and mean depths of some Scottish lakes. <br>
Create vectors, holding the lake's name and all the parameters and build a dataframe called scottish.lakes from the vectors.

```r
loch <- c("Loch Ness", "Loch Lomond", "Loch Morar", "Loch Tay", "Loch Awe", "Loch Maree", "Loch Ericht", "Loch Lochy",
         "Loch Rannoch", "Loch Shiel", "Loch Katrine", "Loch Arkaig", "Loch Shin")
volume <- c(7.45, 2.6, 2.3, 1.6, 1.2, 1.09, 1.08, 1.07, 0.97, 0.79, 0.77, 0.75, 0.35)
area <- c(56, 71, 27, 26.4, 39, 28.6, 18.6, 16, 19, 19.5, 12.4, 16, 22.5)
length <- c(39, 36, 18.8, 23, 41, 20, 23, 16, 15.7, 28, 12.9, 19.3, 27.8)
max_depth <- c(230, 190, 310, 150, 94, 114, 156, 162, 134, 128, 151, 109, 49)
mean_depth <- c(132, 37, 87, 60.6, 32, 38, 57.6, 70, 51, 40, 43.4, 46.5, 15.5)
```

```r
scottish.lakes <- data.frame(loch, volume, area, length, max_depth, mean_depth, stringsAsFactors=FALSE)
```

```r
scottish.lakes
```

```r
str(scottish.lakes)
```

### 1. Evaluate the highest and lowest volume and area lake

```r
max_vol <- scottish.lakes[which.max(scottish.lakes$volume),]
max_area <- scottish.lakes[which.min(scottish.lakes$volume),]
min_vol <- scottish.lakes[which.max(scottish.lakes$area),]
min_area <- scottish.lakes[which.min(scottish.lakes$area),]
```

```r
paste("Lake with highest volume:", max_vol[1,1])
paste("Lake with highest area:", min_vol[1,1])
paste("Lake with lowest volume:", max_area[1,1])
paste("Lake with lowest area:", min_area[1,1])
```

### 2. Order the frame with respect to the area and determine the two largest area lakes

```r
max_two <- scottish.lakes[order(area, decreasing=TRUE),]
```

```r
paste("The two lakes with highest area are", max_two[1,1], "and", max_two[2,1])
```

### 3. By summing up the areas occpupied by the lakes, determine the area of Scotland covered by water

```r
paste("The area of Scotland covered by lakes is", sum(scottish.lakes$area), "km$^2$.")
```

## Exercise 2 - Crude Oil Production

The following CSV file (https://drive.google.com/file/d/1cUQ8d0kDOw2lmzMAHIPglpdcnHy4GgF3/view) contains data on crude oil prices from 1861 to 2019, measured in US dollars per barrel.

```r
library(tibble)
library(ggplot2)
```

### 1. Write R code that is able to read the file and import it in a data frame structure. 
(Hint: before loading the file, open it with a text editor and check its structure).


```r
file = "https://drive.google.com/u/0/uc?id=1cUQ8d0kDOw2lmzMAHIPglpdcnHy4GgF3&export=download"
oil_prices = read.csv(file, quote = "", skip=1, header = T)
head(oil_prices, 10)
```

```r
oil_prices$Code = NULL
oil_prices$Entity = NULL
```

```r
head(oil_prices, 10)
```

### 2. Produce a plot with the Oil price as a function of the year

```r
options(repr.plot.width=10, repr.plot.height=6)
ggplot(oil_prices, aes(x=Year, y=OilPrice)) + geom_line(linetype="solid", color="orchid", size=1) + coord_fixed() + 
ggtitle("Price of crude oil over the years from 1861 to 2019") + ylab("Oil price") + 
scale_x_continuous(breaks = scales::pretty_breaks(n = 15)) + scale_y_continuous(breaks = scales::pretty_breaks(n = 15)) +
theme(plot.title = element_text(color="black", size=14, face="bold.italic"), 
      axis.title.x = element_text(color="black", size=14, face="bold"),
      axis.title.y = element_text(color="black", size=14, face="bold"))
```

### 3. Which is the highest price in history ? When did it occur ?

```r
max_vol <- oil_prices[which.max(oil_prices$OilPrice),]
```

```r
paste("The highest price for crude oil was", round(max_vol$OilPrice, 3), "and it occurred in", max_vol$Year)
```

### 4. Plot the derivative of the curve, simply evaluated with the finite difference formula (forward derivative)

```r
oil_prices$derivative <- c(diff(oil_prices$OilPrice),0)
head(oil_prices,10)
```

```r
options(repr.plot.width=10, repr.plot.height=6)
ggplot(oil_prices[1:158,], aes(x=Year[1:158], y=derivative[1:158])) + 
geom_line(linetype="solid", color="maroon", size=1) + 
ggtitle("Derivative of the price of crude oil over the years from 1861 to 2018") + ylab("Derivative of oil price") + 
xlab("Year") + scale_x_continuous(breaks = scales::pretty_breaks(n = 15)) +
scale_y_continuous(breaks = scales::pretty_breaks(n = 15)) +
theme(plot.title = element_text(color="black", size=14, face="bold.italic"), 
      axis.title.x = element_text(color="black", size=12),
      axis.title.y = element_text(color="black", size=12))
```

# Exercise 3 - World Coal Production

The following CSV file (https://drive.google.com/file/d/10kgRQGKEC81QoVW9kj8goNzzyLcT0xld/view) contains data on the coal production (in TW/hour) for several countries in the World. <br>

### 1. Write R code that is able to read the file and import it in a tibble structure

```r
#file2 = "coal-production-by-country.csv"
file2 = "https://drive.google.com/u/0/uc?id=10kgRQGKEC81QoVW9kj8goNzzyLcT0xld&export=download"
coal_prod = tibble(read.csv(file2, quote = "", header = T, stringsAsFactors=FALSE))
```

I decided to drop the column "Code" since it didn't contain any relevant information. At first, I thought thet the fact that the value of "Code" was NA for geographic areas (such as Africa, Eurasia...) was needed to filter the dataset and take out these regions. However, by looking at the dataset I found out that even some countries didn't have a value for "Code" and so I concluded it wasn't containing relevant information.

```r
coal_prod$Code = NULL
colnames(coal_prod) = c("Entity", "Year", "Prod_TWh")
head(coal_prod, 10)
```

```r
str(coal_prod)
```

### 2. Count the number of countries available in the file and produce a barplot with the number of entries for each country

```r
hist = tibble(as.data.frame(table(coal_prod$Entity)))
colnames(hist) = c("Entity", "Frequency")
str(hist)
```

```r
paste("The number of countries available in the file is", length(hist$Entity))
```

```r
options(repr.plot.width=10, repr.plot.height=30)
ggplot(hist, aes(x=Entity, y=Frequency)) + geom_bar(stat='identity', colour='maroon', fill='maroon') + coord_flip()+
theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
geom_text(aes(label=Frequency, size=0.5), hjust=0, size=3, color='black')
```

#### for the following items select only the years > 1970:

### 3. Selecting only the year after 1970, determine the total integrated production for each country and print the top 5 Countries with highest coal productions

```r
post_1970 <- coal_prod[coal_prod$Year >= 1970, ]
```

```r
int_prod_per_country <- aggregate(post_1970["Prod_TWh"], by=post_1970["Entity"], sum)
int_prod_per_country = int_prod_per_country[order(int_prod_per_country$Prod_TWh, decreasing=TRUE),]
```

```r
paste('The 5 top countries with highest total integrated production of coal are:')
paste(int_prod_per_country[4,1])
paste(int_prod_per_country[7,1])
paste(int_prod_per_country[11,1])
paste(int_prod_per_country[13,1])
paste(int_prod_per_country[14,1])
```

### 4. For the 5 top Countries, create a plot of production as a function of time

```r
tmp = c("China","United States", "Russia", "Australia", "India")
top5 <- post_1970[post_1970$Entity %in% tmp,]
```

```r
options(repr.plot.width=15, repr.plot.height=9)
ggplot(top5, aes(Year, Prod_TWh, group=Entity, colour=Entity)) + geom_point() + geom_line() +
labs(title='Production of coal from 1970 to 2019 for the 5 top countries in the world', x='Year', y='Coal production (TW/h)', col='Country')+ 
scale_x_continuous(breaks = scales::pretty_breaks(n = 15)) + scale_y_continuous(breaks = scales::pretty_breaks(n = 15)) +
theme(plot.title = element_text(color="black", size=18, face="bold.italic"), 
      axis.title.x = element_text(color="black", size=16, face="bold"),
      axis.title.y = element_text(color="black", size=16, face="bold"), 
      axis.ticks.length=unit(.25, "cm"),
      axis.text.x = element_text(face="bold", size=12),
      axis.text.y = element_text(face="bold", size=12))
```

### 5. Generate a plot with the cumulative sum of the World's coal production over the years

```r
tmp = list("World")
world <- post_1970[post_1970$Entity %in% tmp,]
```

```r
cum_sum <- cumsum(world$Prod_TWh)
```

```r
options(repr.plot.width=15, repr.plot.height=6)
ggplot(world, aes(x=Year, y=cum_sum)) + geom_point() + geom_line() +
labs(title="Cumulative sum of the world's coal production from 1970 to 2019", x='Year', y='Coal production (TW/h)')+ 
scale_x_continuous(breaks = scales::pretty_breaks(n = 15)) + scale_y_continuous(breaks = scales::pretty_breaks(n = 15)) +
theme(plot.title = element_text(color="black", size=18, face="bold.italic"), 
      axis.title.x = element_text(color="black", size=16, face="bold"),
      axis.title.y = element_text(color="black", size=16, face="bold"), 
      axis.ticks.length=unit(.25, "cm"),
      axis.text.x = element_text(face="bold", size=12),
      axis.text.y = element_text(face="bold", size=12))
```
